**Oppgave 1**: måler temperaturen i et gitt rom på samme sted, én gang hvert tiende minutt. Tempraturen bør holdes på 20 grader.

```python
T_målt1 = np.array([20.6, 20.4, 20.4, 20.6, 20.4, 20.8, 20.5, 20.5, 20.5, 20.4,
                    20.5, 20.5, 20.5, 20.5, 20.4, 20.4, 20.4, 20.5, 20.3, 20.6]) # 20 første målinger

# (a) finner et std.avvik for målingene
n = len(T_målt1)
m = (1/n) * sum(T_målt1)
mu = np.mean(T_målt1)
sum_til_s = 0
for i in range(len(T_målt1)):
    sum_til_s += (T_målt1[i] - m)**2

s = np.sqrt((1 / (n-1)) * sum_til_s)
print("(a) standardavvik: ", round(s, 4))

# (b) et 95% konfidensintervall for middeltemperaturen i rommet
t = 2.093
mu_minus = m - t * s * np.sqrt(1/n)
mu_pluss = m + t * s * np.sqrt(1/n)

print(f"(b) 95% konfidensintervall: [{round(mu_minus, 2)}, {round(mu_pluss, 2)}]")

# (c) prediksjonsintervall for neste målepunkt: 
mu_neste_minus = m - t * s * np.sqrt(1 + 1/n)
mu_neste_pluss = m + t * s * np.sqrt(1 + 1/n)

print(f"(c) 95% prediksjonsintervall for neste målepunkt: [{round(mu_neste_minus, 2)}, {round(mu_neste_pluss, 2)}]")
```

**Oppgave 3**: R=1kO, presisjon 1%. Antar resistansverdi har en uniform fordeling +-1%

```python
# (a) maksimal/minimal verdi, std.avvik for motstandsverdiene
R = 1000
maks_verdi = R + R*0.01         # b
min_verdi = R - R*0.01          # a
std_avvik = (maks_verdi - min_verdi) / np.sqrt(12)
print(f"(a) maksimal verdi: {int(maks_verdi)}")
print(f"    minimal verdi: {int(min_verdi)}")
print(f"    standard avvik: {round(std_avvik, 2)} \n")

# (b) relative std.avviket:
mu = (maks_verdi + min_verdi) / 2
rel_std_avvik = std_avvik / mu
print(f" (b) relativ std.avvik: {round(rel_std_avvik, 4)}")
print(f"     relativ std.avvik [%]: {round(rel_std_avvik*100, 2)}% \n")

# (c) summering av to uniforme sannsynlighetsfordelinger
print("(c) Svar ligger i innlevering \n")

# (d) relativ std.avvik:
N = 100000
R_500 = 500
dR_500 = 0.01*500
r_values = R_500 + 2*dR_500 * (np.random.rand(N, 2) - 0.5)
r_values_sum = np.sum(r_values, axis=1)
relative_std = np.std(r_values_sum, ddof=1)/np.mean(r_values_sum)

print(f"(d) relativ std.avvik: {round(relative_std, 4)}")
print(f"    relativ std.avvik i %: {round(relative_std*100, 2)}% \n")
print("    Fra resultatene ser vi at vi får en mindre feil når man \n" + 
      "    seriekobler to motstander enn å bruke en motstand, \n" + 
      "    hvis alle har samme presisjon")
```
