### Oppgave 2: Maksimal Relativ Feil (mhp. f)

```python
# c

# for monte carlo simulations
n_random = 10000
tau_vector = (1 + 2 * 0.01 * (np.random.rand(n_random) - 0.5))
d_vector = (1 + 2 * 0.02 * (np.random.rand(n_random) - 0.5))

f_vector = tau_vector / d_vector # leave out c because it is a constant
expected_f = 1
max_error = np.max(np.abs(f_vector - expected_f))/expected_f

print(f"Max relative error: {100*max_error:.2f}%")
```

### Oppgave 3: Maksimal Relativ Feil (mhp. theta)

```python
# c

n_random_numbers = 10000
theta_vector = np.deg2rad([5, 15, 20, 30, 45, 60, 75, 80, 85, 90])
tau_vector = (1 + 2 * 0.01 * (np.random.rand(n_random_numbers) - 0.5))
d_vector = (1 + 2 * 0.02 * (np.random.rand(n_random_numbers) - 0.5))

for theta0 in theta_vector:
    f0 = np.cos(theta0)
    f_vector = f0 * (tau_vector / d_vector)
    remove_outliers = (f_vector <= 1) & (f_vector >= -1)
    f_valid = f_vector[remove_outliers]

    theta = np.arccos(f_valid)

    rel_err = np.abs(theta - theta0) / theta0
    max_error = np.max(rel_err)
    print(f"For angle: {theta0:.2f} rad ({np.rad2deg(theta0):.2f} deg), we have max relative error: {max_error:.4f}")
```
