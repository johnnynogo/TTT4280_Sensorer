```python
# b

T_60_uten = np.array([4.28, 4.28, 4.13, 3.76, 4.14, 4.33, 4.10, 4.21])      # [s]
T_60_med = np.array([3.14, 3.15, 3.92, 3.65, 3.35, 3.89, 3.79, 3.32])       # [s]
n = len(T_60_uten)                                                          # [#]
S_absorbent = 10                                                            # [m^2]
c = 343.4                                                                   # [m/s]
V = 240                                                                     # [m^3]
k = (24 * V * np.log(10)) / (c * S_absorbent)                               # [something]

m_60_med = 1/n * np.sum(T_60_med)
m_60_uten = 1/n * np.sum(T_60_uten)

s_60_med  = np.sqrt(np.sum((T_60_med  - m_60_med )**2) / (n - 1))
s_60_uten = np.sqrt(np.sum((T_60_uten - m_60_uten)**2) / (n - 1))

s_alpha = k * np.sqrt((s_60_med**2 / m_60_med**4) + (s_60_uten**2 / m_60_uten**4))
print(f"Estimated std.dev. for alpha_absorb: {s_alpha:.2}")


print(s_60_med)
print(s_60_uten)
```

```python
# c

m = k * (1/m_60_med - 1/m_60_uten)
print(round(m, 2))

print(m-2.365*s_alpha/np.sqrt(n))
print(m+2.365*s_alpha/np.sqrt(n))
```
